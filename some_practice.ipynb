```python
#large if sales_cost>20


transaction["sales_type"] = np.where(transaction["sales_cost"]> 20, "Large", "Small")
transaction[["sales_cost", "sales_type"]]
```

```python
condition_rules = [transaction["sales_cost"] > 50,
                   transaction["sales_cost"] > 20,
                   transaction["sales_cost"] >10]
outcomes = ["X-Large",
            "Large",
            "Medium"]
transaction["sales_type"] = np.select(condition_rules, outcomes, default = "Small")
transaction["sales_type"]. head(25)
```
