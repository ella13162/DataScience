```python
target = df['SalePrice']
target.head()

# Distribution plot of SalePrice
sns.distplot(target)

# Skewness of SalePrice
print(target.skew())
# Mean (average) of SalePrice
print(target.mean())
# Mode (most frequent value) of SalePrice
# Because SalePrice is positevely skewed, the mode will be lower than the mean/average value
print(target.mode()[0])
```
