#Week 3 Seminar: Plotting in Python
The built-in library in Python called Matplotlib can be used to generate graphs in Python. Within the Matplotlib library there is a specific module called Pyplot which simplifies generation and manipulation of plots in Python.


```
import matplotlib.pyplot as plt   #imports the Pyplot module from the Matplotlib library using a shorthand format called plt
                                  #thus rather than using matplotlib.pyplot we can use its plt alias
```

Pyplot has high level functions such as bar(), plot(), scatter() and pie() that make it easy to generate, bar charts, line charts, scatter plots and pie charts.

##Generating Line Graph
A line graph shows markers connected together in the form of a line.

```
x = [0, 2 ,4, 7, 9, 11, 15]    #creates a list and adds its value to variable x
y = [1, 2, 6, 14, 19, 25, 32]  #creates another list and adds its value to variable y
plt.plot(x,y, 'rd:')           #uses the plot() function within the Pyplot module to plot x values against the y values
                               #the b in "bD-" represents blue colour for markers, D represents diamond shapes and - solid line.
```

We can use the following characters to display the markers and the line in different colours.

We can label the graph's axes and add title using the following lines of code.

```
plt.xlabel("Time (s)", fontsize = 12)               #labels the x-axis using fontsize of 12
plt.ylabel("Cumulative Sales (£)", fontsize = 12)    #labels the y-axis using fontsize of 12
plt.title("Sales over time", fontsize = 14)         #adds title for the graph using fontsize of 14
plt.plot(x,y, 'g*--')                                #plots x values against y values using red colour, point marker and dotted line style
```

##Scatter Plot
A scatter plot shows individual values on cartesian (x,y) plane. Each value is displayed by a marker.

```
playing_chess = [[1,2,3,4,5,5],[10,12,15,20,30,25]] #Creates a list that contains two other lists inside it
watching_tv = [[2,3,4,5,6,4.5],[2,4,7,9,11,14]]     #Creates a list that contains two other lists inside it
plt.xlabel("Hours", fontsize = 12)
plt.ylabel("Age", fontsize = 12)
plt.title("The time a player spends on each activity based on his/her age", fontsize = 12)
plt.scatter(playing_chess[0], playing_chess[1], color = "magenta", marker = "s", label = "Playing Chess") #contents of the list are plotted
plt.scatter(watching_tv[0], watching_tv[1], color = "g", marker = ">", label = "Watching TV")
plt.legend()
```

On the above scatter plot, two sets of values were plotted which are shown in different colours. Additional sets of value can be added using the approach shown.

A scatter plot can be used to show the linear relationship (correlation) between two variables. When showing correlation, a line of best fit can be used to indicate the linear trend.

```
x = (13.5, 11.6, 14.2, 12.1, 13.4, 11.24)
y = (5.6, 8.4, 4.9, 7.8, 8.1, 9.7)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph title")
plt.scatter(x,y)
```

To add a line of best fit, we need to use the numpy function polyfit(). A one dimensional polynomial class called poly1d is created based on the array returned by polyfit().

```
import numpy as np
z = np.polyfit(x,y,1)   #uses the polyfit() function from the numpy library on x and y values
w = np.poly1d(z)        #creates a one dimensional object based on the array returned by polyfit()
plt.scatter(x,y)        #displays the scatter plot
plt.plot(x,w(x), "r")   #generates a set of values, joined by a straight line, for use with x values
```

##Pie Chart
A pie chart divides a circle into sectors, each of which represents a proportion of the whole.

```
countries = ["England", "Scotland", "Wales", "Northern Ireland"]
covid19_cases = [912000, 67000, 54500, 40200]
plt.pie(covid19_cases, labels = countries, autopct = '%1.f%%')
```

##Bar Chart
A bar chart uses horizonal or vertical bars to display categorical data. It is simple to interpret and an effective tool for comparing different categories of data.

```
uk_employment_categories = ("Non-graduates", "Graduates", "Postgraduates")
uk_employment_rate_2019 =  [72, 87.5, 88.7]
index = (1,2,3)
plt.bar(index, uk_employment_rate_2019, tick_label = uk_employment_categories )
plt.title("Percentage of working age employment")
```

The bars can be displayed horizontally by changing plt.bar() to plt.barh().

We can specify the colour for each of the bars.


```
plt.barh(index, uk_employment_rate_2019, tick_label = uk_employment_categories, color = ("red", "blue", "orange"))
plt.title("Percentage of working age employment")
```

### Matplotlib: Histogram
A histogram is a powerful graphical representation used to visualise the distribution of a dataset. It is particularly useful when you need to understand the underlying frequency distribution of a continuous variable. By grouping data into bins or intervals, a histogram displays the number of observations within each bin, helping to identify patterns such as skewness, modality (e.g., unimodal, bimodal), and the presence of outliers. This makes histograms ideal for analysing the spread and central tendency of data, comparing different distributions, and identifying any deviations from a normal distribution. They are commonly used in various fields, including statistics, finance, quality control, and any scenario where it is important to understand the data distribution, such as in exploratory data analysis (EDA) before performing more advanced statistical analyses or machine learning models.

It is worth mentioning that NumPy is an essential library for numerical computing in Python, providing support for arrays, matrices, and a variety of mathematical functions. It excels in handling large datasets efficiently, making it indispensable for scientific computing and data analysis. Its array objects are more efficient and convenient than Python's built-in lists, especially for large-scale computations. By integrating seamlessly with other libraries like SciPy, pandas, and Matplotlib, NumPy forms the foundation of the scientific Python ecosystem, enabling more complex data manipulation and analysis tasks.

```

# Histogram
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Matplotlib: Box Plot

A box plot, also known as a box-and-whisker plot, is a graphical representation used to display the distribution of a dataset based on a five-number summary: minimum, first quartile, median, third quartile, and maximum. It provides a visual summary of data variability, highlighting the central tendency, dispersion, and skewness. The 'box' shows the interquartile range (IQR), with the line inside representing the median, while the 'whiskers' extend to the minimum and maximum values within 1.5 times the IQR from the quartiles. Outliers are plotted as individual points beyond the whiskers. Box plots are particularly useful for comparing distributions between different groups and identifying potential outliers.

```

# Box Plot
data = [np.random.randn(100) for _ in range(4)]

plt.boxplot(data, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Seaborn: Heatmap

Seaborn is a Python visualisation library built on top of Matplotlib, designed to simplify the creation of informative and attractive statistical graphics. It offers a high-level interface for drawing a wide array of plots, such as bar plots, violin plots, pair plots, and heatmaps, with less code and more aesthetic appeal than Matplotlib alone. Seaborn integrates well with pandas data structures, making it particularly convenient for data analysis and visualisation tasks. It excels in handling complex visualisations, including those involving multiple variables, and provides built-in themes for enhancing the visual appeal of charts. Seaborn's ability to manage statistical relationships and its user-friendly interface make it an invaluable tool for data scientists and analysts.

```

# Heatmap
import seaborn as sns
import numpy as np

data = np.random.rand(10, 12)
ax = sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

Seaborn is built on top of Matplotlib and aims to simplify the process of creating attractive and informative statistical graphics. It provides a high-level interface that abstracts much of the complexity of Matplotlib, enabling users to generate complex plots with fewer lines of code. Seaborn excels in handling Pandas DataFrame objects directly, making it easier to work with datasets for statistical analysis. It comes with built-in themes and color palettes to improve the aesthetics of the plots and offers advanced plotting functions for complex visualizations, such as categorical plots, violin plots, and pair plots.

### Seaborn: Pair Plot
Pair plots, created using Seaborn, are a powerful visualisation tool for exploring relationships between multiple variables in a dataset. They generate a matrix of scatter plots, with each plot representing the relationship between a pair of variables, and diagonal plots typically displaying the distribution of each variable. By colour-coding different categories, pair plots can highlight patterns and correlations within the data, making it easier to identify trends, outliers, and interactions among variables. This makes pair plots particularly useful for exploratory data analysis, providing a comprehensive overview of the dataset in a single visualisation.

```

# Pair Plot
import seaborn as sns
import pandas as pd

data = sns.load_dataset('iris')
sns.pairplot(data, hue='species')
plt.title('Pair Plot')
plt.show()
```
